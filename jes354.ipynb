**Data Set:** Number of Trees  

```python
import pandas as pd 
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

pittsTree = pd.read_csv("https://data.wprdc.org/datastore/dump/1515a93c-73e3-4425-9b35-1cd11b2196da", index_col="id", dtype={"id": int})

print(pittsTree)

Quantity = []

trees = {"Neighborhood": pittsTree.loc[:, "neighborhood"]}
df = pd.DataFrame(trees)

for value in df["Neighborhood"]:
    Quantity.append(1)

df["Quantity"] = Quantity

cheese = df.groupby("Neighborhood").sum()


cheese
```

```python
# Metric
maxVal = cheese.max()
print(maxVal)

    
cheese = cheese.sort_values(by=['Quantity'], ascending=False)

cheese["Rating"] = (cheese['Quantity'] / 5073 * 100)

cheeto = {"Rating": cheese.loc[:, "Rating"]}
gf = pd.DataFrame(cheeto)

print(cheese)
print(cheeto)
print(cheese.plot(kind='barh', figsize=(13,20), legend=False))
print(gf.plot(kind='barh', figsize=(10,20), legend=False))
```

**Conclusion:** Based on this dataset, the best neighborhood is Squirrel Hill South. They contain the most trees out of anyone at a total of 5073 trees. That is a lot considering how urban Pittsburgh is. Though it is not exactly suprising that a neighborhood based around a park would edge out the victory here.
